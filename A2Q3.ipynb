```python
### 3a Transition(p)/Prior(pi)

## Read the file into a sting called text
with open('bible.txt', 'r') as myfile:
    text = myfile.read()


from collections import Counter

# Determine transition matrix
pDist = [[0 for x in range(26)] for y in range(26)] 
p = [[0 for x in range(26)] for y in range(26)] 

alph_num = {
    "a": 0,    "b": 1,    "c": 2,    "d": 3,    "e": 4,
    "f": 5,    "g": 6,    "h": 7,    "i": 8,    "j": 9,
    "k": 10,    "l": 11,    "m": 12,    "n": 13,    "o": 14,
    "p": 15,    "q": 16,    "r": 17,    "s": 18,    "t": 19,
    "u": 20,    "v": 21,    "w": 22,    "x": 23,    "y": 24,    "z": 25
}

iterL1 = text[0] #Initialisation
for i in range(1,len(text)):
    iterL2 = text[i]
    
    if (iterL1.isalpha() and iterL2.isalpha()) == True: # Only consider intra-word transitions
        pDist[alph_num[iterL1]][alph_num[iterL2]] += 1
        
    iterL1 = iterL2

totTran = 0 
for i in range(26):
    totTran = sum(pDist[i])
    
    for j in range(26):
        p[i][j] = pDist[i][j]/totTran

        
# Determine vector of prior probabilities 
alph = "abcdefghijklmnopqrstuvwxyz"

priorDistDict = dict(Counter(text))
priorDist = [0 for x in range(26)]
prior = [0 for x in range(26)]

for i in range(26):
    priorDist[i] = priorDistDict[alph[i]]
    
totChar = sum(priorDist)
for i in range(26):
    prior[i] = priorDist[i]/totChar
```
